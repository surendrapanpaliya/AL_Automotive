NumPy (Numerical Python)
Fast, efficient handling of large numerical arrays and mathematical operations.
✅ Key Features:
•	Multi-dimensional arrays (ndarray)
•	Linear algebra, Fourier transform, statistical operations
•	Efficient performance (C-based backend)

```python
import numpy as np
speed_data = np.array([60, 62, 65, 67, 70])

print(speed_data)

average_speed = np.mean(speed_data)

print(average_speed)

```

Pandas
Analyze and manipulate tabular data with DataFrames (like Excel sheets but smarter).

Key Features:
•	Data cleaning, filtering, grouping
•	Time-series handling
•	Integration with CSV/Excel/SQL files


```python
import pandas as pd
df = pd.read_csv('/Users/surendra/gen-ai-jlr/vehicle_data.csv')
print(df.head())
print(df.describe())
print(df.groupby('vehicle_type')['fuel_efficiency'].mean())
```

Matplotlib
Purpose: Basic data visualization library for creating static, animated, and interactive plots.
Key Features:
•	Line, bar, pie charts, histograms
•	Customization of plots (labels, grids, titles)

```python
import matplotlib.pyplot as plt
plt.plot(df['vehicle_type'], df['fuel_efficiency'])
plt.title('Vehicle Fuel Efficiency')
plt.xlabel('vehicle_type')
plt.ylabel('Efficiency (km/l)')
plt.grid(True)
plt.show()
```
